## Python03_Dictionary

```python
thisdict = {
"brand": "Ford",
"model": "Mustang",
"year": 1964
}
print(thisdict)
print(thisdict["year"])
```

```python
print(len(thisdict))
```

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964,
  "year": 2020
}
print(thisdict)
```

```python
thisdict = {
"brand": "Ford",
"electric": False,
"year": 1964,
"colors": ["red", "white", "blue"]
}
print(thisdict)
```

```python
thisdict = dict(name = "John", age = 36, country = "Norway")
print(thisdict)
```

```python
car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
x = car.keys()
print(x)
car["color"] = "white"
print(x)
```

```python
print(car)
```

```python
car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
x = car.values()
print(x)
car["year"] = 2020
print(x)
```

```python
car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
x = car.values()
print(x) #before the change
car["color"] = "red"
print(x)
```

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
if "model" in thisdict:
  print("Yes, 'model' is one of the keys in the thisdict dictionary")
```

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.update({"color": "red"})
print(thisdict)
```

```python
thisdict = {
"brand": "Ford",
"model": "Mustang",
"year": 1964
}
del thisdict["model"]
print(thisdict)
```

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.clear()
print(thisdict)
```

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
for x in thisdict:
  print(x)

for x in thisdict:
  print(thisdict[x])  

for x,y in thisdict.items():
  print(x, y)
```

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
mydict = thisdict.copy()
print(mydict)
```

```python
myfamily = {
"child1" : {
"name" : "Emil",
"year" : 2004
  },
"child2" : {
"name" : "Tobias",
"year" : 2007
  },
"child3" : {
"name" : "Linus",
"year" : 2011
  }
}
print(myfamily)
print(myfamily["child1"])
```

```python
for x, obj in myfamily.items():
  print(x)
for y in obj:
  print(y + ':', obj[y])
```
